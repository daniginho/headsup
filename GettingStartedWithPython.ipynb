# Python3

## Packages for python
Python packages; things like p5 (graphics/gfx package)

We install packages with [pip](https://www.educba.com/python-pip/). 
When we install packages they make a mess of the filesystem.  We also want to make sure that when we copy the code to another machine that we recreate all the code. 

We also want to make sure that each program only has the packages it needs.

You can use a tool called pip to do this.

Install it using 
```sh
sudo apt install python3-pip
# list global packages
pip3 list 
```

These are in the global environment we actually want a simple environment just for our program - a virtual env

```sh
sudo apt install python3-venv
```


So we want to install [pipenv](https://pipenv.pypa.io/en/latest/)...
```sh
sudo pip3 install pipenv
```

## Version of python
As a side there is also a tool called pyenv to manage different versions of python.  



## Creating a new project

```sh
mkdir mycode
cd mycode

# list packages (lots of global rubbish)
pip3 list

#create a pip environment 
pipenv --three

# jump into environment 
pipenv shell

# list packages (only ones we want)
pip3 list

```
